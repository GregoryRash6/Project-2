```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
# File Path
spotify_file = "spotify_data.csv"

# Read CSV
spotify_df = pd.read_csv(spotify_file)

# Keep Only 2017 Dates
spotify_df = spotify_df[spotify_df["Date"] <= "2017-12-31"]

# Show Data
spotify_df
```

```python
# Make a Copy, Keeping Columns Listed Above
spotify_transformed = spotify_df.copy()

# Rename Columns
spotify_transformed = spotify_transformed.rename(columns={"Position": "position",
                                                  "Track Name": "track_name",
                                                  "Artist": "artist",
                                                  "Streams": "streams",
                                                  "URL": "url",
                                                  "Date": "date",
                                                  "Region": "region"
                                                 })
```

```python
# Set Connection String
connection_string = f"postgres:{password}@localhost:5432/spotify_db"

# Set Engine
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm Tables
engine.table_names()
```

```python
# Fill Unemployment Table
spotify_transformed.to_sql(name='spotify', con=engine, if_exists='append', index=False)
```
